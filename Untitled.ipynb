```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df_tracks=pd.read_csv(r"D:\data sets\tracks.csv\tracks.csv")
df_tracks.head()
```

```python
# Checking for null values in the dataset

df_tracks.isnull().sum()
```

```python
# shape of the dataset

df_tracks.shape
```

```python
# Checking dtypes 
df_tracks.info()
```

```python
df_tracks.dropna(inplace=True)
```

```python
# Checking for duplicate rows 
df_tracks.duplicated().sum()
```

```python
df_tracks.shape
```

```python
# Getting the 10 least popular songs on spotify

sorted_df=df_tracks.sort_values("popularity",ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# Top 10 popular songs 

most_popular=df_tracks.query("popularity>90").sort_values("popularity",ascending=False)
most_popular[:10]
```

```python
# Changing the index and converting it into datetime format

df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Converting duration in ms to s 
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop(["duration_ms"],axis=1,inplace=True)
```

```python
df_tracks.duration.head()
```

```python
corr_df_tracks=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(18,8))
heatmap=sns.heatmap(corr_df_tracks,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1)
heatmap.set_title("Corelation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
df_tracks["dates"]=df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```
